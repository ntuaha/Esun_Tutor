### 延伸議題 - 浮點數
首先我們要先知道一件事情，在十進位的0.1 表示的是 $\frac{1}{10}$ ，也就是表示十份中拿一份，所以他會等於$(0.1)_{10}$，但是在二進位的$(0.1)_{2}$表示的事情是$\frac{1}{2}$，這表示的事情是$ (0.1)_{2} = (0.5)_{10}$，所以說如果這個概念有的話，我們就能很快速地理解$(0.11)_{2} = (0.1)_{2} + (0.01)_{2} = \frac{1}{2} + \frac{\frac{1}{2}}{2} = (0.5)_{10} + (0.25)_{10} = (0.75)_{10} $

但這都是從二進位轉換到十進位的時候可以這麼簡單的處理，但要是從十進位的0.1轉換到二進位的表示呢？那就會看見一個無窮無盡的二進位小數，同時要是我們知道電腦在儲存一個數值的時候都只能用固定長度來表示，那就會發生明明我需要用到32個空間來儲存一個十進位小數，但現在我只能用16個空間儲存，那就會導致呈現時候，在電腦的計算精準度過高時，就會明明當初計算的是$\frac{1}{10} $但是卻跑出 0.100000000001 這個怪異數字出來，這是浮點數的特色，也告訴我們在下列兩個情況需要特別留意的:

1. 需要用到等號的場景
2. 需要用到完全精準的場景

這兩個場景在金融很常會發生，所以在開發上就需要特別留意．至於其他國防, 航太等領域我就不熟了...
