## 時間處理
請參考Python的[datetime](https://docs.python.org/2/library/datetime.html),主要是處理:
- 從文字轉成時間 strptime
- 將時間輸出成特定文字 strftime
- 將時間位移到特定時間 timedelta

### 從文字轉成時間

```python
# 將 20170115 轉換成 時間格式
import datetime
datetime.datetime.strptime("20170115","%Y%M%d")
```

### 將時間格式輸出成指定文字格式

```python
# 將 現在時間輸出成 四碼年-月-日 時:分:秒
datetime.datetime.now().strftime("%Y-%m-%d %H:%I:%S")
```

### 取得從現在時間往前一個月
請參考[此連結](http://stackoverflow.com/questions/9724906/python-date-of-the-previous-month)解法:
- find today 取得今日日期
- use that to find the first day of this month. 取得現在這個月的1號
- use timedelta to backup a single day, to the last day of the previous month. 然後往前一天
- print the YYYYMM string you're looking for. 印出年月(格式:YYYYMM)
- 如果需要，可以保留現在的日期，並更改日期

```python
# 取得上一個月的日期, 現在是 20170116  希望得到 20161216
d = datetime.datetime.now()
d.strftime("%Y-%m-%d")
```

```python
now_date = d.day
```

```python
first = d.replace(day=1)
previous_month = first - datetime.timedelta(days=1)
target_date = previous_month.replace(day=now_date)
```

```python
target_date.strftime("%Y-%m-%d")
```

## 爬蟲(1)

### 取得網頁內容
請參考 [requests](http://docs.python-requests.org/en/master/) 說明文件

```python
import requests
r = requests.get("https://www.esunbank.com.tw/bank/personal")
r.text.replace("\r","").replace("\n","")
```

### 進行處理
請參考 [lxml](http://lxml.de/) 說明文件

```python
from lxml import etree
from io import StringIO
```

### 將網頁轉換成DOM tree

```python
tree = etree.parse(StringIO(r.text.replace("\r","").replace("\n","")), etree.HTMLParser())
```

### 取得玉山公告標題

```python
print(tree.xpath('//*[@id="mainform"]/div[7]/div/div/div[1]/h2')[0].text)
```

### 取得玉山公告新聞連結

```python
# 找新聞其中一個連結的id, 然後用 .. 回到他的母元素之後, 往下找所有的 a
links = tree.xpath('//*[@id="layout_0_maincontent_1_announcement_0_RepAnnocement_Li1_0"]/..//a')
for link in links:
    print(link.text)
```
