```python
url = 'http://www.society.taichung.gov.tw/section/index-1.asp?Parser=99,16,257,,,,3807,589,,,,42,,3'
careCenter = pd.read_html(url)[1]
```

```python
careCenter.columns = careCenter.ix[1]
```

```python
careCenter= careCenter[2:]
```

```python
careCenter.columns
```

```python
#careCenter.drop('行政區域',axis=1,inplace=True)
```

```python
careCenter['行政區域'] = (careCenter[careCenter.columns[0]].fillna(method='ffill'))
```

```python
careCenter
```

```python
import re
def 養護(t):
    match = re.findall('養護(\d+)', t)
    if match:
        return int(match[0])
    else:
        return None
```

```python
careCenter['目前收容人數(養護)'] = careCenter['目前收容人數'].apply(養護)
careCenter['床型/床數(養護)'] = careCenter['床型/床數'].apply(養護)
```

```python
careCenter.head()
```

```python
import matplotlib.pyplot as plt
plt.figure()
careCenter[['床型/床數(養護)','床型/床數(養護)','行政區域']].plot()
plt.show()
```
