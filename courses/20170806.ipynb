```python
# votes 的 DataFrame
v = pd.DataFrame(data=list(data['votes'].values()))
v.head()
```

```python
v[v.source=='3']
```

```python
# 建立對照表
IT = pd.DataFrame(data=data['ticket'])
IT.columns = ['source_name','source']
IT
```

```python
# 只抽出必要的欄位
core_data = v.loc[:,["name","source","vote"]]
core_data.head()
```

```python
d1 = pd.merge(core_data,IT,on='source')
d1.head()
```

```python
IT.columns = ['vote_name','vote']
d2 = pd.merge(d1,IT,on='vote')
d2.head()
```

```python
import numpy as np
```

```python
pd.pivot_table(d2, values=['name'],index=['source_name'],columns=['vote_name'], aggfunc=np.count_nonzero,margins=True)
```
